```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


df = pd.read_csv('Top_Highest_Openings.csv')
print(df)
```

```python
# Pie Chart
gross_by_distributor = df.groupby('Distributor')['Total Gross'].sum().reset_index()

plt.figure(figsize=(8, 6))
plt.pie(gross_by_distributor['Total Gross'], labels=gross_by_distributor['Distributor'], autopct='%1.1f%%')
plt.title('Total Gross Revenue by Distributor')
plt.show()
```

```python
# Bar Chart
ax = df[['Opening','Total Gross']].plot(kind='bar', title ="Gross vs Opening", figsize=(15, 10), legend=True, fontsize=12)
ax.set_xlabel("Film", fontsize=12)
ax.set_ylabel("Billions", fontsize=12)
plt.show()
```

```python
# Line Graph
plt.plot(df["Theaters"], df["Average"])
plt.title("Theater Earnings")
plt.xlabel("Number of Theaters Filmed Was Shown In")
plt.ylabel("Average Earning Per Theater")
plt.show()
```
